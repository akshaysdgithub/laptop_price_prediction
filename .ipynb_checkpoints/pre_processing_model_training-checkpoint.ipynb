### Univariate Analysis

#### External Features:  Inches, Weight_Kg, Resolution, Touchscreen

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(15,8))

sns.boxplot(df['Inches'], ax=axs[0][0])
axs[0][0].set_title("Inches")
sns.histplot(df['Inches'], ax=axs[0][1])

sns.boxplot(df['Weight_Kg'], ax=axs[1][0])
axs[1][0].set_title('Weight_Kg')
sns.histplot(df['Weight_Kg'], ax=axs[1][1])

plt.show()
```

```python
l = sorted(df['Resolution'].value_counts().index.to_list())
print(l)
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(10,8))

sns.barplot(x=df['Price'],y=df['Resolution'],orient='h',order=l,ax=axs[0])

sns.barplot(x=df['Price'],y=df['Touchscreen'],orient='h',ax=axs[1])

plt.show()
```

#### Hardware Features: Ram, Cpu_Brand, Cpu_Proc, Gpu_Type, Cpu_Ghz, SSD, HDD, Hybrid, Flash_Storage

```python
l = ['Ram', 'Cpu_GHz', 'SSD', 'HDD', 'Hybrid', 'Flash_Storage']
fig, axs = plt.subplots(nrows=len(l), ncols=2, figsize=(20,30))

for i in range(len(l)):
    name = l[i]
    sns.boxplot(df[name], ax=axs[i][0])
    axs[i][0].set_title(name)
    sns.histplot(df[name], ax=axs[i][1])

plt.show()
```

#### Software Features: OpSys

```python
sns.countplot(y=df['OpSys'], orient='h')
plt.show()
```

#### Brand Features: Company, TypeName

```python
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(15,10))

sns.countplot(x=df['Company'], orient='h', ax=axs[0])
sns.countplot(x=df['TypeName'], orient='h', ax=axs[1])

plt.show()
```
